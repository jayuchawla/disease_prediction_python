```python
import pandas as pd
import csv
from collections import defaultdict
```

```python
#split and append only disease name
def return_list(disease):
    disease_list = []
    match = disease.replace('^','_').split('_')
    ctr = 1
    for group in match:
        if ctr%2==0:
            disease_list.append(group)
        ctr = ctr + 1

    return disease_list
```

```python
disease_list = []
with open("dis1.csv") as csvfile:
    reader = csv.reader(csvfile)
    
    disease=""
    weight = 0
    disease_list = []
    dict_wt = {}
    ####value is of type list
    dict_=defaultdict(list)
    
    for row in reader:
        #print(row)
        if row[0]!="\xc2\xa0" and row[0]!="":
            disease = row[0]
            disease_list = return_list(disease)
            weight = row[1]

        if row[2]!="\xc2\xa0" and row[2]!="":
            symptom_list = return_list(row[2])
            ####appending disease wise symptoms in dict_ and disease wise weights in dict_wt
            for d in disease_list:
                for s in symptom_list:
                    dict_[d].append(s)
                dict_wt[d] = weight
print(dict_)
print('---------------------')
print(dict_wt)
print('======================')
```

```python
###cleaning data by filling each column of each row by repective disease and weights.
with open("dis.csv","w") as csvfile:
    writer = csv.writer(csvfile)
    for key,values in dict_.items():
        for v in values:
            key = str.encode(key).decode('utf-8')
            writer.writerow([key,v,dict_wt[key]])
```

# Naive Bayes on cleaned dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
```

```python
####use k-fold cross validation
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
from sklearn.metrics import accuracy_score
##multinomial Naive Bayes also called Conditional probablity
##if I am interested in finding frequency of a particular word in a text file there multinomial Naive Bayes is used.
##it is used to evaluate combined probablity suppose you have 6 people, u need to find combined prob of 1 having O blood group, 2 having A bg, 2 having B bg, 1 having AB bg
##you do this by p(x1=1 and x2=2 and x3=2 and x4=1) = (6!/1!*2!*2!*1!)*(p1^1)*(p2^2)*(p3^2)*(p4^1)
mnb = MultinomialNB()
mnb.fit(x_train, y_train)
y_pr = mnb.predict(x_test)
print(accuracy_score(y_test,y_pr))
```

```python
##It can't work on unseen data because it has never seen that disease before. Also, there is only one point
#for each disease and hence no point for this. So we need to train the model entirely.
mnb_tot = MultinomialNB()
mnb_tot = mnb_tot.fit(x, y)
mnb_tot.score(x,y)
```

```python
disease_pred = mnb_tot.predict(x)
disease_real = y.values
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

con_mat = confusion_matrix(disease_real,disease_pred)
print(len(con_mat))
print('confusion_mat:')
print(con_mat)
print('accuracy: ')
print(accuracy_score(disease_real,disease_pred))
print('classify: ')
print(classification_report(disease_real,disease_pred))
```

```python
#print those not match to real
for i in range(0, len(disease_real)):
    if disease_pred[i]!=disease_real[i]:
        print ('Pred: {0} Actual:{1}'.format(disease_pred[i], disease_real[i]))
```
